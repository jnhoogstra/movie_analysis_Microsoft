```python
df_rotten_movies = pd.read_csv("data/rotten_tomatoes_movies.csv")
df_rotten_reviews = pd.read_csv("data/rotten_tomatoes_critic_reviews.csv")
df_title_basics = pd.read_csv("data/title.basics.csv")
df_tmdb = pd.read_csv("data/tmdb.movies.csv")
df_budgets = pd.read_csv("data/tn.movie_budgets.csv")
```

```python
df_budgets.info()
```

```python
df_budgets["production_budget"] = df_budgets["production_budget"].str.replace("$", "")
df_budgets["domestic_gross"] = df_budgets["domestic_gross"].str.replace("$", "")
df_budgets["worldwide_gross"] = df_budgets["worldwide_gross"].str.replace("$", "")

df_budgets["production_budget"] = df_budgets["production_budget"].str.replace(",", "")
df_budgets["domestic_gross"] = df_budgets["domestic_gross"].str.replace(",", "")
df_budgets["worldwide_gross"] = df_budgets["worldwide_gross"].str.replace(",", "")
```

```python
df_budgets["production_budget"] = pd.to_numeric(df_budgets["production_budget"])
df_budgets["domestic_gross"] = pd.to_numeric(df_budgets["domestic_gross"])
df_budgets["worldwide_gross"] = pd.to_numeric(df_budgets["worldwide_gross"])

df_budgets.info()
```

```python
df_budgets["profit"] = df_budgets.domestic_gross + df_budgets.worldwide_gross - df_budgets.production_budget

df_budgets.head()
```
